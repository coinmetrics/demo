```python
import os
import pandas as pd

api_key = os.getenv('CM_API_KEY') # Obtain the API key from environment variable.
```

```python
from coinmetrics.api_client import CoinMetricsClient

client = CoinMetricsClient(api_key)
```

```python
my_markets = client.reference_data_markets(
    ).to_dataframe()
my_markets
```

```python
# kraken futures
df = my_markets.loc[
    (my_markets.exchange=='kraken') &
    (my_markets.type=='future') &
    (my_markets.base.isin(['btc', 'eth']))
]
df
```

```python
availability = client.catalog_market_candles_v2(
    markets=list(df.market),
).to_dataframe().merge(df)
availability.loc[availability.frequency=='1m', ['market', 'expiration', 'min_time', 'max_time', 'base']].sort_values(by=['expiration'], ascending=False)
```

`min_time` is the datetime when that value is first available. see below for a way to convert this to Eastern.

```python
# Create a DatetimeIndex for every day until 2035
all_days = pd.date_range(start='2025-01-01', end='2035-12-31')

# Convert to DataFrame for easier filtering and grouping
df = pd.DataFrame({'date': all_days})

df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month

# Filter to only include Fridays (where weekday=4)
fridays_only = df[df['date'].dt.weekday == 4]

# Group by year, month and get the last Friday of each month
last_fridays = fridays_only.groupby(['year', 'month']).last()
last_fridays
```

```python
# predict the expiration of a month in the future...
month = 10
year = 2025

last_fridays.loc[year, month]
```
